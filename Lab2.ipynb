```python
from sqlalchemy import create_engine
from sqlalchemy.engine import URL
from urllib.parse import unquote

server_name   = "localhost"
database_name = "Lab2"

connection_string = f"DRIVER=ODBC Driver 17 for SQL Server;SERVER={server_name};DATABASE={database_name};TrustServerCertificate=yes;UID=sa;PWD=123456;"
url_string        = URL.create("mssql+pyodbc", query={"odbc_connect": connection_string})

print('Connecting to database using URL string:')
unquoted_url = unquote(str(url_string))
print(unquoted_url, '\n')

try:    
    engine = create_engine(url_string)
    with engine.connect() as connection:
        print(f'Successfully connected to {database_name}!')
except Exception as e:
    print('Error while connecting to database:\n')
    print(e)
```

```python
import pandas as pd

df_books = pd.read_csv("Books.csv", sep=';')
df_books
```

```python
df_books.to_sql('Books', con=engine, if_exists='replace', index=False)
```

```python
df_inventory = pd.read_csv("Inventory.csv", sep=';')
df_inventory
```

```python
import random

# Get the number of rows in the dataframe
num_rows = len(df_inventory)

# Generate a list of random numbers between 0 and 120
random_numbers = [random.randint(0, 120) for _ in range(num_rows)]

# Replace the values in the "Amount of Books" column
df_inventory["Amount of books"] = random_numbers

# Set at least 10 books with 0 amount
num_zero_books = min(10, num_rows)
zero_books_indices = random.sample(range(num_rows), num_zero_books)
df_inventory.loc[zero_books_indices, "Amount of books"] = 0

# Print the updated dataframe
df_inventory
```

```python
df_inventory.to_sql('Inventory', con=engine, if_exists='replace', index=False)
```

```python
# Merge df_books and df_stores to get all combinations of ISBN13 and StoreID
df_combinations = df_books.merge(df_stores, how='cross')

# Explode the ISBN13 column to create one row for each ISBN13
df_combinations_exploded = df_combinations.explode('ISBN13')

# Rename the column to 'ISBN13' for consistency
df_combinations_exploded.rename(columns={'ISBN13': 'ISBN13'}, inplace=True)

new_df = df_combinations_exploded.iloc[:, [0, 5]]
new_df
```

```python
import random

# Set the seed for reproducibility
random.seed(42)

# Generate random numbers between 0 and 120
random_numbers = [random.randint(0, 120) for _ in range(len(new_df))]

# Ensure that at least 25 rows have a value of 0
num_zeros = 25
zero_indices = random.sample(range(len(new_df)), num_zeros)
for index in zero_indices:
    random_numbers[index] = 0

# Add the new column to the DataFrame
new_df['Random Number'] = random_numbers
new_df
```

```python
new_df.rename(columns={'Random Number': 'Amount of books'}, inplace=True)
```

```python
df_final_inventory = new_df.copy()
df_final_inventory = df_final_inventory.iloc[:, [0, 2, 1]]
df_final_inventory
```

```python
df_final_inventory.to_sql('Inventory', con=engine, if_exists='replace', index=False)
```
